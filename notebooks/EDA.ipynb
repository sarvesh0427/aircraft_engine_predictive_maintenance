```python
columns = ['engine_id', 'cycle'] + \
          [f'op_setting_{i}' for i in range(1, 4)] + \
          [f'sensor_{i}' for i in range(1, 22)]
```

```python
df = pd.read_csv(
    r'..\data\raw\train_FD001.txt',
    sep=' ',
    header=None,
    names=columns
)
```
